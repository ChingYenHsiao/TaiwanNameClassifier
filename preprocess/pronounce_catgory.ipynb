```python
Not_spand_list =['參見','異體字','異體']

def get_all_define(word,spand_times,finished_term_list):

    defines=''
    for hete in word['heteronyms']:
        for define in hete['definitions']:
            defines+=define['def']+'\n'
            #print(word['title']+':'+define['def'])
            if spand_times==1 and len(define['def'])<8 and '參見' not in  define['def'] :
                for term in Term_list: 
                    if term in finished_term_list:
                        continue
                    TF = define['def'].count(term)
                    if TF>0:
                        #print(term+'spand:\n')
                        
                        finished_term_list.append(term)
                        defines+=get_all_define( moe_data_dict[term],0,finished_term_list)
                        
            #「」連結的展開                   
            if 'example' in define or 'link' in define or '參見' in define['def'] or '異體字' in define['def']: 
                d=''
                if '參見' in define['def'] or '異體字' in define['def']:
                    d+=define['def']
                if 'example' in define:
                    d+=define['example'][0]
                if 'link' in define:
                    d+=define['link'][0]
                #print(define['example'][0])
                up = '「'
                up_case_list = [pos for pos, char in enumerate(d) if char == up]
                down = '」'
                down_case_list =  [pos for pos, char in enumerate(d) if char == down]

                for i in range( len(up_case_list) ):
                    term = d[up_case_list[i]+1 :down_case_list[i] ]
                    if term=='參見':
                        continue
                    defines+=term+'\n'
                    if term in moe_data_dict and term not in finished_term_list : 

                        #print(term+'[]:\n')
                        finished_term_list.append(term)
                        defines+=get_all_define( moe_data_dict[term],0,finished_term_list)
    return defines
finished_term_list = ['麗']
print(get_all_define(moe_data_dict['麗'],1,finished_term_list))
```

```python
m_list = ['a','o','e','i','ia','io','ie','u','ua','uo','ü','üe','er','ê'
         'ai','ei','ao','ou','iao','iou','uai','uei',
         'an','ian','uan','üan','en','in','uen','ün','anɡ','iang','uang','enɡ','ing','uenɡ','onɡ','iong']
m_list.sort(key = lambda s: len(s))
m_list.reverse()
print(m_list)
```

```python
max_list = []
for m in m_list:
    if 'a' in m:
        for a in a_l:
            max_list.append( m[:m.index('a')]+a +m[m.index('a')+1:] )

    if 'e' in m:
        for e in e_l:
            max_list.append( m[:m.index('e')]+e +m[m.index('e')+1:] )
    if 'i' in m:
        for i in i_l:
            max_list.append( m[:m.index('i')]+i +m[m.index('i')+1:] )
    if 'o' in m:
        for o in o_l:
            max_list.append( m[:m.index('o')]+o +m[m.index('o')+1:] )
    if 'u' in m:
        for u in u_l:
            max_list.append( m[:m.index('u')]+u +m[m.index('u')+1:] )
    
print(max_list)
```
