# Use WikiCorpus as w2v dataset

```python
# -*- coding: utf-8 -*-
import logging
import sys
from gensim.corpora import WikiCorpus
```

```python
def wiki_extract():
    logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
    wiki_corpus = WikiCorpus('./Wiki/zhwiki-20170920-pages-articles.xml.bz2', dictionary={})
    texts_num = 0
    
    with open("./Wiki/wiki_texts.txt",'w',encoding='utf-8') as output:
        for text in wiki_corpus.get_texts():
            output.write(' '.join(text) + '\n')
            texts_num += 1
            if texts_num % 10000 == 0:
                logging.info("已處理 %d 篇文章" % texts_num)
wiki_extract()
```

```python
from mylib.is_chinese import is_chinese
```

```python
def clean_text(infile,outfile):
    with open(infile,'r') as File:
        out_file = open(outfile,'w')
        for line in File.readlines():
            l = ''
            for c in line:
                if c==' ':
                    l+=c
                else: 
                    if is_chinese(c)==True:
                        l+=c
            l+='\n'
            for i in range(7):
                l= l.replace('  ',' ')
            out_file.write(l)  
        out_file.close()
```
