```python
# -*- coding: utf-8 -*-

#簡轉繁用 
#https://pypi.python.org/pypi/hanziconv/0.2.1
from hanziconv import HanziConv

#import bs4
import sys
import os
import re
import time
import json
#import mechanize
import requests
from bs4 import BeautifulSoup
```

```python
import pickle
```

```python
with open('./data_output/miss_word_list.txt', 'rb') as handle:
    miss_word_list = pickle.loads(handle.read())
```

```python
miss_word_list
```

```python
specail_word_dict = {}
```

```python
for word in miss_word_list:
    #url = "http://zidian.wenku1.com/%E6%B9%8B/"
    url = "http://zidian.wenku1.com/"+word+"/"
    time.sleep(1)
    response = requests.get(url)
    
    if response.status_code != 200:
        print (response.status_code)
        continue
    #response.encoding = 'big5'
    response.encoding='utf-8' 
    #print(response.text)
    soup = BeautifulSoup(response.text, 'lxml')

    articles = soup.find_all('td')
    specail_word_dict[ word ] = {}

    for index,s in  enumerate(articles):
        s = str(s)
        #print(s)
        if index ==0:
            #print(s)
            define =  s[s.index('是：')+3:  s.index( '\" class')].replace(' ',"")
            define = HanziConv.toTraditional(define)
            specail_word_dict[word]['def'] = define
            #print( define)
        if '</a>' in s:
            #print(index,s[ s.index('">')+2 :s.index('</a>') ])
            if index ==3:
                specail_word_dict[word]['pinyin'] = s[ s.index('">')+2 :s.index('</a>') ]
            if index ==5:
                specail_word_dict[word]['radical'] = s[ s.index('">')+2 :s.index('</a>') ]
            if index ==7:
                specail_word_dict[word]['stroke_count'] = s[ s.index('">')+2 :s.index('</a>') ]
            if index ==25:
                specail_word_dict[word]['Five_elements'] = s[ s.index('">')+2 :s.index('</a>') ]
    print(word,specail_word_dict[word])        
   
```
